```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
df['mean compactness']
# df.head()
# df[:100]
# df[99:100]
# df.iloc[[0, 1]]  # Returns the first two rows
# df.iloc[0, 1] # first row & second column
# df.iloc[[0, 2], [1, 3]]  # Returns specific rows and columns
```

```python
row_count,column_count = df.shape
print(f"No of rows in the dataset: {row_count}")
print(f"No of columns in the dataset: {column_count}")
print(f"Each row represent a sample or datapoint & each column represents a feature or attribute associated with the corresponding sample")
```

```python
print(f"Total no of benign cases: {sum(cancer.target)}")
print(f"Total no of malignant cases: {row_count - sum(cancer.target)}")
```

```python
import matplotlib.pyplot as plt

benign_df = df[df['target']==1]
malignant_df = df[df['target']==0]

plt.scatter(benign_df['mean compactness'],benign_df['mean perimeter'],label='Benign',color='blue')
plt.scatter(malignant_df['mean compactness'],malignant_df['mean perimeter'],label='Malignant',color='red')
plt.xlabel('Mean Compactness')
plt.ylabel('Mean Perimeter')
plt.title('Scatter Plot(Mean Compactness & Mean Perimeter): Benign vs Malignant')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['diagnosis'] = df['target'].map({1:'Benign',0:'Malignant'})

sns.set_theme(style="whitegrid")
joint_plot = sns.jointplot(data=df, x="mean compactness", y="mean perimeter", hue="diagnosis", palette={"Benign": "blue", "Malignant": "red"})
joint_plot.fig.suptitle('Joint Plot(Mean Compactness & Mean Perimeter): Benign vs Malignant')
joint_plot.set_axis_labels('Mean Compactness','Mean Perimeter')
plt.legend(title='Diagnosis',loc='upper right')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
features = ['mean texture', 'mean perimeter', 'mean compactness', 'mean radius']
scatter_matrix (df[features],figsize=(12,12),color=['red' if label==0 else 'green' for label in df['target']])
plt.suptitle('Pairwise scatterplot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['diagnosis'] = df['target'].map({1:'Benign',0:'Malignant'})
features = ['mean texture', 'mean perimeter', 'mean compactness', 'mean radius', 'diagnosis']
sns.set_theme(style="whitegrid")

joint_plot = sns.pairplot(data=df[features], hue="diagnosis", palette={"Benign": "blue", "Malignant": "red"})
joint_plot.fig.suptitle('Pair Plot: Benign vs Malignant')
plt.legend(title='Diagnosis',loc='upper right')
plt.show()
```
